Function and a table of (x,y) values

函数与(x,y)值表

```python
x = 5
y = 4*x + 3 # The function

print(x,",",y)

print("x \t y")

for x in range(11): # 打印0-10的数
    y = 4*x + 3  # The function
    print(x, "\t", y)

```

Defning a Python function

定义 Python 函数

```python
def f(x):
    y = 4*x + 3
    return y

print(5, ",", f(5))

# Next, a loop
for x in range(11):
    print(x, "\t", f(x))
```

Graph and table of (x,y) values

图表和(x,y)值表格

```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
ax.set_xlim(xmin, xmax) # window size
ax.set_ylim(ymin, ymax)
# 绘制轴
ax.axhline(y=0, color='b')
ax.axvline(x=0, color='b')

print("x \t y")
for x in range(xmin, xmax+1):
    y = 0.5*x + 1
    ax.plot([x],[y], 'ro')
    print(x,"\t",y) # 打印x,y值

plt.show() # 展示图像
```

Use an array as inputs

使用数组作为输入

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)  # 生成一系列的x值，points有范围决定，动态生成，包括终点

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

y = 2*x +1
plt.plot(x,y, 'pink')

plt.show()
```

More ways to customize a graph

更多自定义图表的方式

##### using the variable axs for multiple Axes
* fig, axs = plt.subplots(2, 2)

##### using tuple unpacking for multiple Axes
* fig, (ax1, ax2) = plt.subplots(1, 2)

* fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    The names ax and pluralized axs are preferred over axes
because for the latter it's not clear if it refers to a single
~.axes.Axes instance or a collection of these.

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots(figsize=(8,6)) # 创建对象，并设置图像大小
ax.set_xlim(xmin, xmax)# window size
ax.set_ylim(ymin, ymax)

ax.axhline(y=0, color='b') # blue x axis
ax.axvline(x=0, color='b') # blue y axis

ax.set_xlabel("x values") #设置标签
ax.set_ylabel("y values")
ax.set_title("Some Graph")  # 设置标题
ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 1))  # x轴的间隔
ax.set_yticks(np.arange(ymin, ymax, 1)) # y轴的间隔

y = 2*x +1
ax.plot(x,y, label='y=2x+1')
ax.plot([4],[6], 'ro', label='point')
ax.plot(x,3*x, label='steeper line')
ax.legend() # 显示图例
plt.show()
```

# 斜率-slope

```python
x1 = 1
y1 = 7
x2 = 2
y2 = 10

slope = (y2 - y1) / (x2 - x1)

print("slope = ", slope)
```

```python
x1 = 632 #1
y1 = 84 #7
x2 = 1800 #2
y2 = 8 #10

# The slope is "m"
m = (y2 - y1) / (x2 - x1)  # 计算斜率

# The y intercept is "b"
b = y1 - m*x1  # 计算截距
# The full equation
print("y = ", m, "x + ", b)
print(f"y = {m}x + {b}") # 推荐
```

```python
import matplotlib.pyplot as plt

x1 = 2
y1 = 3
x2 = 6
y2 = 8

# Develop the equation y = mx + b
m = (y2 - y1) / (x2 - x1)
b = y1 - m*x1
print(f"y = {m}x + {b}")

# For the graph
xmin = -10
xmax = 10
ymin = -20
ymax = 20

# For the line on the graph
y3 = m*xmin + b  # 带入点计算y值
y4 = m*xmax + b

# Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Plot the linear function as a red line
plt.plot([xmin,xmax],[y3,y4],'ro') # 就是两个点，不会做直线
plt.plot([xmin,xmax],[y3,y4],'r') # 会绘制两点直线
plt.show()


```

Solve an equation

求解方程

```python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = 10*x - 1000

solution = solve(eq,x)
print(solution)
print(type(solution))
print("x = ", solution[0]) # 可能有多个解
```
